```python
import numpy as np

line= np.array(lines)
def connect_and_remove_parallel(
    lines: np.ndarray,
    connect_threshold=15,
    overlap_threshold=8
) -> np.ndarray:
    """
    lines: ndarray of shape (N, 1, 4)
    connect_threshold : for snapping endpoints
    overlap_threshold : for removing nearby parallel duplicates
    """

    L = lines.reshape(-1, 4).astype(int).copy()
    n = len(L)

    # ----------------------------
    # STEP 1: CONNECT ENDPOINTS
    # ----------------------------
    for i in range(n):
        x1, y1, x2, y2 = L[i]

        for j in range(n):
            if i == j:
                continue

            a1, b1, a2, b2 = L[j]

            # Vertical lines
            if x1 == x2 and a1 == a2 and abs(x1 - a1) <= overlap_threshold:
                if abs(y2 - b1) <= connect_threshold:
                    L[j][1] = y2
                if abs(y1 - b2) <= connect_threshold:
                    L[j][3] = y1

            # Horizontal lines
            if y1 == y2 and b1 == b2 and abs(y1 - b1) <= overlap_threshold:
                if abs(x2 - a1) <= connect_threshold:
                    L[j][0] = x2
                if abs(x1 - a2) <= connect_threshold:
                    L[j][2] = x1

    # ----------------------------
    # STEP 2: REMOVE PARALLEL CLOSE DUPLICATES
    # ----------------------------
    keep = np.ones(len(L), dtype=bool)

    for i in range(len(L)):
        if not keep[i]:
            continue

        x1, y1, x2, y2 = L[i]

        for j in range(i + 1, len(L)):
            if not keep[j]:
                continue

            a1, b1, a2, b2 = L[j]

            # ---- Horizontal parallel ----
            if y1 == y2 and b1 == b2:
                if abs(y1 - b1) <= overlap_threshold:
                    # Check X overlap
                    if not (x2 < a1 or a2 < x1):
                        keep[j] = False

            # ---- Vertical parallel ----
            if x1 == x2 and a1 == a2:
                if abs(x1 - a1) <= overlap_threshold:
                    # Check Y overlap
                    if not (y2 < b1 or b2 < y1):
                        keep[j] = False

    L = L[keep]

    return L.reshape(-1, 1, 4)

linen= connect_and_remove_parallel(line, connect_threshold=15, overlap_threshold=10)
```
